**ANALISANDO UMA TABELA DE VENDAS COM A BIBLIOTECA PANDAS**

```python
import pandas as pd
```

```python
vendas_df = pd.read_excel('Vendas.xlsx')
display(vendas_df)
```

```python
produto = vendas_df[['Produto', 'ID Loja', 'Quantidade']]
display(produto.loc[1:20])
```

```python
display(vendas_df[['ID Loja', 'Valor Final']].loc[1:20])
```

```python
display(vendas_df.groupby('ID Loja').sum())
```

**FATURAMENTO POR LOJA**

```python
faturamento = vendas_df[['ID Loja', 'Valor Final']].groupby('ID Loja').sum()
display(faturamento)
```

**QUANTIDADE DE PRODUTOS VENDIDOS POR LOJA**

```python
qtd_vendas = vendas_df[['ID Loja', 'Quantidade']].groupby('ID Loja').sum()
display(qtd_vendas)
```

**TICKET MÉDIO DOS PRODUTO EM CADA LOJA**

```python
ticket_medio = (faturamento['Valor Final'] / qtd_vendas['Quantidade']).to_frame()
ticket_medio = ticket_medio.rename(columns={0: 'Ticket Médio'}) # Renomeando a coluna
display(ticket_medio)
```
